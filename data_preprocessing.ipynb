```python
# write function for making list of row numbers from image folders
import os
def getnum(path): 
    files = os.listdir(path) 
    img_num = [] 
    for file in files:
        number = int(file[-9:-4]) 
        img_num.append(number)  
    return img_num
```

```python
# function for labelling eachs entry corresponding to their class
def label(data, num_cleandata, labelclean, num_debrisdata, labeldebris):
    data.loc[:]['Class'][data.loc[:]['Class'].index.isin(num_cleandata)] = labelclean 
    data.loc[:]['Class'][data.loc[:]['Class'].index.isin(num_debrisdata)] = labeldebris 
    return data
```
